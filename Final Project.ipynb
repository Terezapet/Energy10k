```python
import re
import nltk
from nltk.corpus import stopwords

# Download NLTK resources
nltk.download('stopwords')

# Define preprocessing function
def preprocess_text(text):
    # Remove numbers and punctuation except for hyphens
    text = re.sub(r'[^a-zA-Z\s-]', '', text)  # Keep only letters, spaces, and hyphens
    return text.lower()

# Function to create a CRE measure for a given SEC filing (document)
def calculate_cre(text, ccv_terms, gpv_terms):
    """
    Calculate the Climate Regulatory Exposure (CRE) measure for a firm's SEC filing.
    
    Args:
    - text: The preprocessed text of the SEC filing.
    - ccv_terms: A list of climate-related terms (CCV).
    - gpv_terms: A list of government policy-related terms (GPV).
    
    Returns:
    - cre: The computed Climate Regulatory Exposure (CRE) score (1 if both terms found, 0 otherwise).
    """
    # Preprocess the text
    processed_text = preprocess_text(text)
    
    # Initialize CRE score
    cre_score = 0
    
    # Check if any term in CCV dictionary is present in the text
    ccv_found = any(ccv_term in processed_text for ccv_term in ccv_terms)
    
    # Check if any term in GPV dictionary is present in the text
    gpv_found = any(gpv_term in processed_text for gpv_term in gpv_terms)
    
    # If both climate-related and government policy-related terms are found, assign a CRE score
    if ccv_found and gpv_found:
        cre_score = 1
    
    return cre_score

# Updated dictionaries

# Climate-related terms (CCV dictionary)
ccv_terms = [
    'climate change', 'climate', 'global warming', 'clean air act', 'ghg', 'rps', 
    'weather', 'rain', 'emission', 'carbon', 'carbon dioxide', 'carbon tax', 'cap-and-trade', 
    'sustainable', 'renewable', 'green'
]

# Government-related terms (GPV dictionary)
gpv_terms = [
    'congress', 'government regulation', 'political risk', 'congressional', 'government approval', 
    'politics', 'debt ceiling', 'government debt', 'price constraint', 'federal', 'government deficit', 
    'price control', 'federal funds', 'government intervention', 'price restriction', 'fiscal imbalance', 
    'law', 'regulation', 'government-approved', 'legislation', 'regulatory compliance', 'government-sponsored', 
    'legislative', 'regulatory delay', 'governmental', 'legislatory', 'reimbursement', 'government program', 
    'political', 'subsidy', 'government program', 'political', 'subsidies'
]

# Example document (SEC filing text)
example_text = """
    cap-and-trade government-sponsored
"""

# Calculate the CRE for the example document
cre = calculate_cre(example_text, ccv_terms, gpv_terms)
print(f"CRE score for the document: {cre}")
```

```python
# Apply the calculate_cre function to each row in the 'text' column and create a new 'CRE_score' column
labeled_270['CRE_score'] = labeled_270['chunk_text'].apply(lambda x: calculate_cre(x, ccv_terms, gpv_terms))

# Print the DataFrame with the new CRE_score column
print(labeled_270[['chunk_text', 'CRE_score']].tail(50))  # Display the first few rows with the CRE score
```
